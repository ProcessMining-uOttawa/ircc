```python
import pandas as pd
```

```python
df = pd.read_csv("ircc_uOttawa.csv")
```

```python
def get_unique_values(df, attr_name):
    df[attr_name] = df[attr_name].astype(str)
    values = df[attr_name].sort_values().unique()
    print("\n".join(values))
```

```python
get_unique_values(df, 'act_upd_by')
```

```python
get_unique_values(df, 'activity_id')
```

```python
get_unique_values(df, 'group')
```

```python
get_unique_values(df, "cor")
```

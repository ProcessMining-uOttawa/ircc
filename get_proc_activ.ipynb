```python
import pandas as pd
```

```python
def print_unique_activities(name, path):
    print("-", name)
    
    df = pd.read_csv(path)
    # remove subprocess end events
    df = df[~ df['concept:name'].str.endswith(" [end]")]
    # rename subprocess start events
    df['concept:name'] = df['concept:name'].str.replace(r"(.*) (\[start\]|\[end\])", lambda m: f"{m[1]} [subprocess]", regex=True)
    print("\n".join(df['concept:name'].unique()))
    
    print()
```

```python
import os

subfolders = os.listdir("lifecycles")
subfolders.sort()
for f in subfolders:
    p = os.path.join("lifecycles", f)
    
    if os.path.isdir(p):
        print(f"-- Level {f[5]}\n")
        
        p = os.path.join(p, "logs")
        subfolders2 = os.listdir(p)
        subfolders2.sort()
        for f2 in subfolders2:
            fn, _ = os.path.splitext(f2)
            print_unique_activities(fn, os.path.join(p, f2))
            
    print("")
```
